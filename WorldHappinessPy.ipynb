```python
# Dependencies and Set up
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# File (CSV)
data_file = "happiness.csv"
data_file = " un_data.csv"

```

```python
# Load the CSV file into a Pandas DataFrame
un_data_df = pd.read_csv("un_data.csv")

# Display sample data
un_data_df


```

```python
# Load the CSV file into a Pandas DataFrame
happiness_data_df = pd.read_csv("happiness.csv")

# Display sample data
happiness_data_df
```

```python
merged_df = un_data_df.merge(happiness_data_df, how= "inner", left_on="country", right_on="Country")
merged_df
```

```python
from scipy.stats import linregress
x = merged_df["Population in thousands (2017)"]
y = merged_df["Happiness.Score"]

(slope, intercept, r, p, stderr) = linregress(x, y)
print("Pvalue", p)
print("Rvalue", r)

```

```python
x = merged_df["GDP per capita (current US$)"]
y = merged_df["Happiness.Score"]

(slope, intercept, r, p, stderr) = linregress(x, y)
print("Pvalue", p)
print("Rvalue", r)
reg_line = (slope * x) + intercept
plt.scatter(y, x)
plt.plot(y, x,  color="r")
```

```python
#dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd 
import scipy.stats as st
import numpy as np
import ipympl

#study data files
happiness_path = "happiness.csv"


#read the csv files
happiness_df = pd.read_csv(happiness_path)

#display data
happiness_df

```

```python
#Generate Bar Chart Top 5 Countries
Country = ["Norway, Denmark, Iceland, Switzerland, Finland"]
Happiness_Score = [7.537, 7.522, 7.504, 7.494, 7.469]
x_axis = np.arange(len(Happiness_Score))
tick_locations = [value for value in x_axis]
plt.bar(x_axis, Happiness_Score, color="blue", alpha=0.75, align="center")
plt.xticks(tick_locations, ['Norway', 'Denmark', 'Iceland', 'Switzerland', 'Finland'])
plt.xlim(-0.7, len(x_axis)-0.25)
plt.ylim(7.25,7.6)
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.show()
```

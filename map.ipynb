```python
#Dependencies
import hvplot.pandas
import pandas as pd
```

```python
#Import 2017 Happiness Report as a dataFrame
happiness_df = pd.read_csv('happiness.csv')
```

```python
#Import coordinates in longitude and latitudes of all countries as a dataFrame
coordinates_df = pd.read_csv('coordinates.csv')
```

```python
#Merge the two dataframes together
happiess_location_df = happiness_df.merge(coordinates_df, left_on='Country', right_on='country')
```

```python
#Clean up and drop all unecessary columns in the merged dataframe
happy_loc_clean_df = happiess_location_df.drop(['Happiness.Rank', 'Whisker.high', 'Whisker.low', 'Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.', 'Freedom', 'Generosity', 'Trust..Government.Corruption.', 'Dystopia.Residual', 'country', 'usa_state_code', 'usa_state_latitude', 'usa_state_longitude', 'usa_state'], axis=1)
```

```python
#Create a Map Plot of all countries in the 2017 Hapiness Report according to their longitude and latitude and sized by their happines score
map_plot_1 = happy_loc_clean_df.hvplot.points(
    "longitude",
    "latitude",
    geo = True,
    tiles = "EsriImagery",
    frame_width = 800,
    frame_height = 600,
    size = "Happiness.Score" * 100,
    color = "Country"
)
map_plot_1
```

```python
#Create a Map Plot of all countries in the 2017 Hapiness Report according to their longitude and latitude and sized by their happines score
map_plot_2 = happy_loc_clean_df.hvplot.points(
    "longitude",
    "latitude",
    geo = True,
    tiles = "OSM",
    frame_width = 800,
    frame_height = 600,
    size = "Happiness.Score" * 100,
    color = "Country"
)
map_plot_2
```
